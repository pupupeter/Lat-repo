```python

#要問chatgpt的時候 從csv檔中用記事本打開後複製放入再問
#1該補習班'school'欄位中  國立大學與非國立大學的比例如何知道? 
import pandas as pd
df = pd.read_csv('223.csv')
counts = df.groupby('school')['school'].count()#把school的地方先收集起來做總和 groupby是用來把東西集合
nacount=(df['school'].str.contains('國立')).sum()#把school中有'國立'的字收集起來做總和#str是什麼要記下
nonnacount=len(df) - nacount#剩下為非國立的#len是長度(總數)
```

```python
nacountratio = nacount / len(df)
nonnacountratio = nonnacount / len(df)
print('國立大學比例：{:.2f}%'.format(nacountratio*100))#format是把算出來的東西放入前面的{:.2f}%
print('非國立大學比例：{:.2f}%'.format(nonnacountratio*100))
```

```python
#2.該補習班  如何知道 學校 '新莊' 班級 的數A 哪一個班級考得比較好?
```

```python
import pandas as pd
df = pd.read_csv('223.csv')
df['數A'] = df['數A'].str.extract('(\d+)', expand=False)#extract('(\d+)', expand=False 是什麼意思
df = df.fillna('NA')
df = df[df['學校班級'].str.contains('新莊')]
df['數A'] = pd.to_numeric(df['數A'], errors='coerce')#errors 是什麼 to_numeric 是什麼
最好班級 = df[df['數A'] == df['數A'].max()]['學校班級'].unique() # 
print('數A成績最好的班級:', 最好班級)
```

```python
#3.該補習班 school中 非國立大學 多由錄取管道中的哪種方式錄取?
df = pd.read_csv('223.csv')
filtered = df.query("school.str.contains('國立') == False")#用query讀出來然後把school中有國立的剔掉
schoolfiltered = filtered.groupby('入取管道').size()#.size()是什麼要問一下!!!
result = schoolfiltered.idxmax()#idxmax是幫我找出那個schoolfiltered的非國立大學中數出入取管道最多次的，就是找出最多次的字
print("非國立大學多由{}方式入取".format(result))
```

```python
#4.該補習班 學生的國文成績 達到頂標以上的有幾人
import pandas as pd
df = pd.read_csv('223.csv')
df = df.fillna('NA')#要把沒資料的剔掉用fillna，在裡面填NA，這樣就不會連沒有資料的都算入而有異常了
頂標 = df[df['國文'].str.contains('\(頂')].shape[0]#.shape()是什麼要問一下!!!
print("國文成績中達到頂標的人數為{}人".format(頂標))
```

```python
#5.該補習班 "新莊" 班級中  入取管道為 申請入學的 多為什麼school
import pandas as pd #要逐一解釋 因為我有重複的式子，要解釋如何排除複製狀況!!!
df = pd.read_csv('223.csv')
filtered_df = df.loc[(df['學校班級'].str.contains('新莊')) & (df['入取管道'] == '申請'), :].copy()#這裡
filtered_df.loc[:, 'school'] = filtered_df['school'].str.split(' ').str[0]#這裡 
school_count = filtered_df['school'].value_counts()
most_common_school = school_count.idxmax()
print(most_common_school)
```

```python
#6.該補習班 學生的英文成績的平均以及其標準差?
import pandas as pd
df = pd.read_csv("223.csv")
df["英文成績的數值"] = df["英文"].str.extract(r'(\d+)').astype(float)#astype 要看是什麼
mean = df["英文成績的數值"].dropna().mean()#dropna是什麼:刪缺失值、mean 是什麼:平均
std = df["英文成績的數值"].dropna().std()#std 是什麼 標準差
print("英文成績平均值：", mean)
print("英文成績標準差：", std)
```

```python
#7.該補習班 入取為國立學校 最多的 高中學校是哪一所?
df = pd.read_csv("223.csv")
國立大學 = df.loc[df["school"].str.contains("國立", na=False)].copy()
#cannot mask with non-boolean array containing NA / NaN values 的解決----na=False 然後有出現SettingWithCopyWarning 為防複製嫌疑用.copy() 不然換方法
國立大學["學校"] = 國立大學["學校班級"].str.extract(r'(.+?)\d')#(r'(.+?)\d')意思就是我在班級時，抓字串到數字前
df = df.fillna('NA')
國立大學counts = 國立大學.groupby("學校").size()
mostfrequent = 國立大學counts.idxmax()
print("school為國立學校最多的高中學校是：", mostfrequent)
```

```python
#8.該補習班 學生的英文與英聽的相關係數為多少?
import pandas as pd

df = pd.read_csv('223.csv', index_col=0)#index_col 第一列索引
df["英文成績的數值"] = df["英文"].str.extract(r'(\d+)').astype(float)#astype
df['英聽'] = df['英聽'].map({'A': 4, 'B': 3, 'C': 2, 'F': 1})
df.fillna(0, inplace=True)
df["英文成績的數值"] = pd.to_numeric(df["英文成績的數值"], errors='coerce')
corr = df["英文成績的數值"].corr(df['英聽'], method='pearson')
corr
```

```python
#9.該補習班 非國立大學中多以哪所居多
import pandas as pd
df = pd.read_csv("223.csv")
非國立大學 = df.loc[df["school"].str.contains("國立", na=True)].copy()
非國立大學["學校"] = 非國立大學["學校班級"].str.extract(r'(.+?)\d')
df = df.fillna('NA')
非國立大學counts = 非國立大學.groupby("學校").size()#size()
mostfrequent = 非國立大學counts.idxmax()
print("school為非國立學校最多的高中學校是：", mostfrequent,'高中')
```
