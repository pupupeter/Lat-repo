```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_excel("2017_PM25.xlsx")

x = df[['WIND_SPEED','TEMP','RAIN','PH_RAIN','Humidity','RAIN_COND']]
y = df['PM25']

model = LinearRegression()
model.fit(x,y)

coef = model.coef_
intercept = model.intercept_

print(f"回歸方程式為: PM25 = {coef[0]:.2f} * WIND_SPEED + {coef[1]:.2f} * TEMP + {coef[2]:.2f} * RAIN + {coef[3]:.2f} * PH_RAIN + {coef[4]:.2f} * Humidity + {coef[5]:.2f} * RAIN_COND + {intercept:.2f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("2017_PM25.csv")

variables = ["SO2","CO","O3","PM10" ,"PM25","Nox","NO" ,"NO2","THC","NMHC","CH4"]

subset_data = data[variables]

correction_matrix = subset_data.corr()

sns.clustermap(correction_matrix, cmap ='coolwarm', annot=True, fmt='.2f')

plt.title('Clustermap of Variables')

plt.show()
```

```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.read_csv("Tainan_198407-201811.csv")

x = df[['SO2','CO','CO2','O3','Nox','NO','NO2']]
y = df['PM25']
Z= df['PM10']
model = LinearRegression()
model.fit(x,y)
model.fit(x,Z)
coef = model.coef_
intercept = model.intercept_

print(f"回歸方程式為: PM25 = {coef[0]:.2f} * SO2 + {coef[1]:.2f} * CO + {coef[2]:.2f} * CO2 + {coef[3]:.2f} *O3 + {coef[4]:.2f} * Nox + {coef[5]:.2f} * NO +{coef[6]:.2f}*NO2+ {intercept:.2f}")
print(f"回歸方程式為: PM10 = {coef[0]:.2f} * SO2 + {coef[1]:.2f} * CO + {coef[2]:.2f} * CO2 + {coef[3]:.2f} *O3 + {coef[4]:.2f} * Nox + {coef[5]:.2f} * NO +{coef[6]:.2f}*NO2+ {intercept:.2f}")
```
