```python
#變異數越大，表示數據點之間的差異越大，反之亦然 ，指離散程度越大
#假設我們想要比較不同種類的肥料對植物生長的影響。我們從不同的田地中隨機選擇了10個植物，
#每個植物分別用不同種類的肥料進行施肥。在植物生長的過程中，我們記錄了它們的高度。
#現在，我們希望利用變異數分析比較不同種類的肥料對植物生長的影響是否有統計上的差異。
#根據變異數分析的基本假定，我們需要確認以下三點：
#常態分布：我們需要確認每個肥料種類的植物高度數據是否符合正態分布，也就是數據的分佈是否對稱。
#變異數同質性：我們需要確認不同種類的肥料對植物高度的影響是否相同，也就是每個組的數據分散程度是否相同。
#獨立性：我們需要確保每個樣本之間是獨立的，也就是每個植物的生長不會受到其他植物或外界因素的干擾。


#ANOVA用於比較三個或更多組的平均值是否有統計上的差異。
#ANOVA假設各組的數據是獨立、隨機抽樣的，而且服從正態分布。
#ANOVA計算總變異、組間變異和組內變異。總變異等於所有數據的變異。組間變異等於組平均值之間的差異。組內變異等於每個組內部數據點之間的差異。
#通過計算F值，判斷組間變異和組內變異是否有意義。如果F值大於臨界值，則組間差異在統計上是有意義的。
#如果ANOVA結果顯示組之間存在差異，進一步進行事後比較以確定哪些組之間存在差異。

#組內離均差平方和(within sum of squares, SSW)：各組母群體中，所抽出的樣本觀察值與該組樣本平均值之差值的平方總和。 
#組間離均差平方和(between sum of squares, SSB)：各組樣本平均數與總平均數之差值的平方總和。 
#SSB與SSW各自除以自由度(組間自由度為組數 k - 1；組內自由度為 樣本總數n - 組數k)，所得為 組間均方(between groups mean square, MSB) 與 組內均方(within groups mean square, MSW)。 
#最後可利用檢定統計量 F值 = MSB / MSW 
```

```python
#假設你想要比較三種不同肥料對植物生長的效果。你將隨機選擇三個相同大小的區域，並將每個區域分配一種不同的肥料。然後，在每個區域中隨機選擇相同數量的植物，並在同一環境條件下種植。
#在種植一段時間後，你測量了每個區域中的植物生長情況。你得到了以下數據：
#區域 1：肥料 A：10, 12, 14, 13, 11；平均值：12 
#區域 2：肥料 B：9, 11, 8, 10, 12；平均值：10 
#區域 3：肥料 C：15, 13, 16, 14, 12；平均值：14
#為了計算ANOVA，你需要先計算每個組的平均值和整體平均值。
#整體平均值：(10+12+14+13+11+9+11+8+10+12+15+13+16+14+12)/15 = 12
#組間變異數（組間平方和）：((12-12)² x 5) + ((10-12)² x 5) + ((14-12)² x 5) = 20 
#組內變異數（組內平方和）：((10-12)² + (12-12)² + (14-12)² + (13-12)² + (11-12)²) x 3 + ((9-10)² + (11-10)² + (8-10)² + (10-10)² + (12-10)²) x 3 + ((15-14)² + (13-14)² + (16-14)² + (14-14)² + (12-14)²) x 3 = 40
#組間平均差平方和：組間變異數 / (組數 - 1) = 20 / 2 = 10
#組內平均差平方和：組內變異數 / (總樣本數 - 組數) = 40 / 12 = 3.33
#F值：組間平均差平方和 / 組內平均差平方和 = 10 / 3.33 = 3
#即 各組樣本數(ex:區域1的樣本數)x(組平均-全部組別的平均)的平方---> 組間變異數
#即 (一組內各樣本-該組平均)的平方全加起來再x一組內有幾個樣本+另一組的  
```

```python
import pandas as pd
import pingouin as pg
df=pd.read_csv('TPRMOE2018.csv')
print(df.info())
```

```python
wide_format=df[['No','cohnesion1','cohnesion2','cohnesion3']]
wide_format.sort_values('No').head(10)
```

```python
#melt用法---->資料轉置
long_format=pd.melt(wide_format,id_vars='No',
                   value_vars=['cohnesion1','cohnesion2','cohnesion3'],
                   var_name='cohnesion',value_name='cohnvalue')
long_format.sort_values('No').head(10)
```

```python
rmaov=pg.rm_anova(dv='cohnvalue',within='cohnesion',
                 subject='No',correction=False,data=long_format,
                 detailed=True,effsize='ng2')
rmaov
#F值越大越顯著 越小越不顯著 跟P值是對稱的
```

```python
import researchpy as rp
from scipy.stats import ttest_rel
co1=df['cohnesion1']
co2=df['cohnesion2']
co3=df['cohnesion3']
t1=ttest_rel(co1,co2)
t2=ttest_rel(co1,co3)
t3=ttest_rel(co2,co3)
print(rp.summary_cont(df[['cohnesion1','cohnesion2','cohnesion3']]).round(3))
print(t1)
print(t2)
print(t3)

#3>1 2>1  比較2者間差異 要事後比較兩組間差異  (p值)
```

```python
import pingouin as pg
from pingouin import pairwise_ttests
import researchpy as rp
import pandas as pd
df=pd.read_csv('python2011nsc.csv')
print(df.columns)
print(df.info())
```

```python
rp.summary_cont(df['pl'].groupby(df['Race'])).round(3)
```

```python
import scipy.stats as stats
stats.levene(df['pl'][df['Race']==1],
             df['pl'][df['Race']==2],
             df['pl'][df['Race']==3],
             center ='mean')
```

```python
aov=pg.anova(dv='pl',between='Race',data=df,detailed=True ).round(3)
aov
```

```python
G=df['Race'].unique()
args=[]
for i in list(G):
    args.append(df[df['Race']==i]['pl'])
```

```python
from scipy import stats
stats.levene(*args)
```

```python
from scipy import stats
stats.f_oneway(*args)
```
