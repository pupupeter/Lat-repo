```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
df=pd.DataFrame({"x" : [137.97,104.50,100.00,124.32,79.20,99.00,124.00,114.00,106.69,138.05,53.75,46.91,68.00,63.02,81.26,86.21],
                 "y" :[145.00,110.00,93.00,116.00,65.32,104.00,118.00,91.00,62.00,133.00,51.00,45.00,78.50,69.65,75.69,95.30]})
a=df[["x"]]
b=df[["y"]]
model=LinearRegression()
model.fit(a,b)
coef=model.coef_
intercept=model.intercept_
print("線性方程式為: y={}*x+{}".format(coef[0][0],intercept[0]))
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

df = pd.DataFrame({"x": [137.97, 104.50, 100.00, 124.32, 79.20, 99.00, 124.00, 114.00, 106.69, 138.05, 53.75, 46.91, 68.00, 63.02, 81.26, 86.21],
                   "y": [145.00, 110.00, 93.00, 116.00, 65.32, 104.00, 118.00, 91.00, 62.00, 133.00, 51.00, 45.00, 78.50, 69.65, 75.69, 95.30]})

a = df[["x"]]
b = df[["y"]]

model = LinearRegression()
model.fit(a, b)

coef = model.coef_[0][0]
intercept = model.intercept_[0]

print("線性方程式為: y = {:.2f} * x + {:.2f}".format(coef, intercept))
```

```python
from sklearn.datasets import load_boston
import pandas as pd
import statsmodels.api as sm

# 載入波士頓房價數據集
boston = load_boston()

# 將資料轉換成 DataFrame 格式
df = pd.DataFrame(boston.data, columns=boston.feature_names)

# 將標籤（房價）加入 DataFrame
df['MEDV'] = boston.target

# 簡單線性回歸（以 RM 為自變數，MEDV 為應變數）
# 多元線性回歸（以所有變數為自變數，MEDV 為應變數）
X = df.drop('MEDV', axis=1)
y = df['MEDV']
X = sm.add_constant(X)  # 加入截距項
model = sm.OLS(y, X).fit()  # 建立多元線性回歸模型
print(model.summary())  # 輸出迴歸結果
 # 建立簡單線性回歸模型
print(model.summary())  # 輸出迴歸結果
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
import pandas as pd
import statsmodels.api as sm

# 載入波士頓房價數據集
boston = load_boston()

# 將資料轉換成 DataFrame 格式
df = pd.DataFrame(boston.data, columns=boston.feature_names)

# 將標籤（房價）加入 DataFrame
df['MEDV'] = boston.target

# 簡單線性回歸（以 CRIM 為自變數，MEDV 為應變數）
X = df['CRIM']
y = df['MEDV']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

# 繪製散點圖和迴歸直線
sns.regplot(x='CRIM', y='MEDV', data=df)
plt.title('Simple Linear Regression')
plt.xlabel('CRIM')
plt.ylabel('MEDV')

# 輸出迴歸方程式
print(model.summary())
print('y=24.03x-0.41')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取資料集，假設已經有 season 欄位
df = pd.read_csv('SeoulBikeData.csv', encoding='gbk')

# 計算每個季節的平均腳踏車租用次數
season_mean = df.groupby('Seasons')['Rented Bike Count'].mean()

# 繪製直方圖
plt.bar(season_mean.index, season_mean.values)
plt.xlabel('季節')
plt.ylabel('平均腳踏車租用次數')
plt.show()

# 找出租用次數最高的季節
max_season = season_mean.idxmax()
print('租用次數最高的季節為', max_season)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
data=pd.read_csv("SeoulBikeData.csv")
variables=['Rented Bike Count', 'Seasons']
subset_data=data[variables]
model=ols("Q('Rented Bike Count')~C(Seasons)",data=subset_data).fit()
anova_table=sm.stats.anova_lm(model)
print("ANOVA分析摘要表:")
print(anova_table)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.stats.multicomp as mc
data=pd.read_csv("SeoulBikeData.csv")
variables=['Rented Bike Count', 'Seasons']
subset_data=data[variables]
model=ols("Q('Rented Bike Count')~C(Seasons)",data=subset_data).fit()
anova_table=sm.stats.anova_lm(model)
posthoc=mc.MultiComparison(subset_data["Rented Bike Count"],subset_data['Seasons'])
posthoc_results=posthoc.tukeyhsd()
print("事後比較結果:")
print(posthoc_results)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("SeoulBikeData.csv")
data["Date"] = pd.to_datetime(data['Date'], format="%d/%m/%Y", dayfirst=True)
data['Year'] = data['Date'].dt.year
data["Month"] = data['Date'].dt.month
filtered_data = data[(data['Year'] >= 2017) & (data['Year'] <= 2018)]
rented_bikes_by_month = filtered_data.groupby(["Year", "Month"])["Rented Bike Count"].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(rented_bikes_by_month["Year"].astype(str) + '-' + rented_bikes_by_month['Month'].astype(str),
         rented_bikes_by_month["Rented Bike Count"])
plt.xlabel('Year-Month')
plt.ylabel('Rented Bike Count')
plt.title("Trend of rented bike count (2017-2018)")
plt.xticks(rotation=45)
plt.show()
```
