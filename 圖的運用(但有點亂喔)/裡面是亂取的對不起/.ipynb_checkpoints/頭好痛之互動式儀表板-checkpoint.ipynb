```python
import pandas as pd
df=pd.read_csv('python2011nsc.csv')
pd.set_option('display.max_rows',None)
df
```

```python

print(df.describe)
```

```python

print('嗨阿明')

print(df[['pl','al']].describe())#pL靜態休閒 AL動態休閒
```

```python
#單一樣本t檢定
from scipy import stats
print(df['Grade'].mean(),df['Grade'].std())
stats.ttest_1samp(df['Grade'],70)
#statistic=-6.747xxxxxx 代表差平均6.747xxxxx分 這叫做統計量t值
#pvalue=3.955xxxxxxxe-11 他小於 0.05 代表學生成績與期望70分 有顯著差異   ## p>0.05   無顯著差異 P<0.05 有顯著差異 P稱為 虛無假設
```

```python
#Levene 變異數同質性檢定
import scipy.stats as stats
stats.levene(df['pl'][df['Gender']==1],
             df['pl'][df['Gender']==0],
             center='mean')
#0.24>0.05  沒顯著差異------>男性跟女性在靜態休閒上變異數same
#center取值取決於資料分布型態，常態分佈mean ,若非常態分布 取中位數median
```

```python
#執行獨立樣本t檢定
import scipy.stats as stats
group1=df['pl'][df['Gender']==1]
group2=df['pl'][df['Gender']==0]
print(group1.count())
print(group1.mean()) 
print(group1.std())
print(group2.count())
print(group2.mean())
print(group2.std())
stats.ttest_ind(group1,group2,equal_var=True)
```

```python
import researchpy as rp
rp.ttest(group1=df['pl'][df['Gender']==1],group1_name='Male',
         group2=df['pl'][df['Gender']==0],group2_name='Female')
```

```python
#成對樣本t檢定
import pandas as pd
df1=pd.read_csv('TPRMOE2018.csv')

print(df1.info())
```

```python
#成對樣本t檢定#問(團體凝聚力前測與後測)
from scipy.stats import ttest_rel
co1=df1['cohnesion1']
co2=df1['cohnesion2']
print(co1.mean()) 
print(co2.mean()) 
print(co1.std())
print(co2.std())
ttest_rel(co1,co2)
```

```python
import pandas as pd
import scipy.stats as stats
df=pd.read_csv('python2011nsc.csv')


g1=df['achievement'][df['Gender']==1]
g2=df['achievement'][df['Gender']==0]
print('個數','平均數','標準差')
print('男生',g1.count(),g1.mean(),g1.std(),) 
print('女生',g2.count(),g2.mean(),g2.std(),) 
stats.ttest_ind(g1,g2,equal_var=True)
```

```python
#researchpy的p.ttest()
import researchpy as rp
rp.ttest(group1=df['achievement'][df['Gender']==1],group1_name='Male',
         group2=df['achievement'][df['Gender']==0],group2_name='Female')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics as st
import seaborn as sns
from sklearn.datasets  import load_iris
iris=load_iris()
x=iris.data
y=iris.target
df_data=pd.DataFrame(data=np.c_[iris['data'],iris['target']],
                     columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

df_data
```

```python
#直方圖
df_data.hist(alpha=0.6,layout=(3,3),figsize=(12,8),bins=10)
plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=4)
fig.set_size_inches(15,4)
sns.histplot(df_data['SepalLengthCm'][:],ax=axes[0],kde=True)
sns.histplot(df_data['SepalWidthCm'][:],ax=axes[1],kde=True)
sns.histplot(df_data['PetalLengthCm'][:],ax=axes[2],kde=True)
sns.histplot(df_data['PetalWidthCm'][:],ax=axes[3],kde=True)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df_data,figsize=(10,10),color='b',diagonal='kde')
```

```python
sns.pairplot(df_data, hue='Species',height=2,diag_kind='kde')
```

```python
corr=df_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']].corr()
plt.figure(figsize=(8,8))
sns.heatmap(corr,square=True, annot=True, cmap = 'Greys')#cmap type 有很多顏色可以改
```

```python
sns.lmplot('PetalLengthCm','PetalWidthCm',hue='Species',data=df_data,fit_reg=False,legend=False)
plt.legend(title='Species',loc='upper right',labels=['Iris-Setosa','Iris-Versicolour','Iris-Virginica'])
```
