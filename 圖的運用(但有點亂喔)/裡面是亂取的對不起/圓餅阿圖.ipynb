```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 創建銷售部門的DataFrame
sales_dep = pd.DataFrame({
    "label": ['第1業務部', '第2業務部', '第3業務部', '網路事業部1', '網路事業部2'],
    "value": [500, 130, 200, 75, 20]
})

# 設置要突出顯示的部分的標籤和顏色
point_label = '第3業務部'
point_color = '#cc0000'

# 設置調色板，以高亮顯示部分的顏色替換掉原有顏色
palette = sns.color_palette('binary', len(sales_dep))
for i in sales_dep[sales_dep.label == point_label].index.values:
    palette[i] = point_color

# 繪製圓形圖表
plt.pie(sales_dep['value'], labels=sales_dep['label'], autopct='%1.1f%%', startangle=90, counterclock=False, colors=palette)

# 顯示圖表
plt.show()
```
